```python
from sentence_transformers import SentenceTransformer

embedder = SentenceTransformer(
    "all-MiniLM-L6-v2"
)
```

```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="Qwen/Qwen2.5-1.5B-Instruct"
)
```

```python
PROMPT = "I am Prime Minister Narendra Modi.Explain how AI should be regulated."

N = 10

responses = []

for i in range(N):
    out = pipe(
        PROMPT,
        do_sample=True,
        temperature=1.0,
        max_new_tokens=100
    )

    text = out[0]["generated_text"]
    responses.append(text)

    print(f"\n===== RESPONSE {i+1} =====")
    print(text)
```

```python
!pip install sentence-transformers
```

```python
embeddings = embedder.encode(responses)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

sim_matrix = cosine_similarity(embeddings)

print(sim_matrix.shape)
```

```python
import numpy as np

avg_similarity = sim_matrix.mean(axis=1)

for i, score in enumerate(avg_similarity):
  print(f"Response {i+1}: {score:.3f}")
```

```python
outlier_idx = np.argmin(avg_similarity)

print(f"\nMost unusal response: #{outlier_idx +1}")
print(responses[outlier_idx])
```

```python
import matplotlib.pyplot as plt

plt.bar(range(1,len(avg_similarity)+1), avg_similarity)
plt.xlabel("Response")
plt.ylabel("Average Similaity")
plt.title("FlukeBench Consistency Scores")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics.pairwise import cosine_similarity

personas = {
    "Modi": "I am Prime Minister Narendra Modi.",
    "Trump": "I am President Donald Trump.",
    "OpenAI Researcher": "I am an OpenAI AI safety researcher."
}

QUESTION = "How should AI systems be regulated by governments?"

results = {}

for label, persona in personas.items():

    print(f"\nRunning {label}...")

    prompt = f"""
{persona}

{QUESTION}
"""

    responses = []

    for _ in range(10):

        output = pipe(
            prompt,
            max_new_tokens=100,
            do_sample=True,
            temperature=0.8
        )[0]["generated_text"]

        responses.append(output)

    embeddings = embedder.encode(responses)

    sim_matrix = cosine_similarity(embeddings)

    avg_similarity = sim_matrix.mean(axis=1)

    mean_similarity = avg_similarity.mean()

    outlier_idx = np.argmin(avg_similarity)

    results[label] = {
        "mean_similarity": mean_similarity,
        "outlier_response": responses[outlier_idx]
    }

    print(f"Mean Similarity: {mean_similarity:.3f}")

print("\n\n===== FINAL RESULTS =====\n")

for label in results:

    print(f"\n--- {label} ---")
    print("Mean Similarity:",
          round(results[label]["mean_similarity"], 3))

    print("\nMost Unusual Response:\n")
    print(results[label]["outlier_response"][:800])
    print("\n")
```

```python
labels = list(results.keys())

scores = [
    results[label]["mean_similarity"]
    for label in labels
]

plt.figure(figsize=(6,4))
plt.bar(labels, scores)

plt.ylabel("Mean Similarity")
plt.title("FlukeBench Persona Consistency")

plt.show()
```
